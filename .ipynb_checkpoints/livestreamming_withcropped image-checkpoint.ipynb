```python
import cv2
 
#img = cv2.imread('nature.jpg', cv2.IMREAD_UNCHANGED)
 
#print('Original Dimensions : ',photo.shape)
cap = cv2.VideoCapture(0)

while True:
    ret, photo = cap.read()
    scale_percent = 60 # percent of original size
    width = int(photo.shape[1] * scale_percent / 200)
    height = int(photo.shape[0] * scale_percent / 200)
    dim = (width, height)
    # resize imagr
    resized = cv2.resize(photo, dim, interpolation = cv2.INTER_AREA)
    print('Resized Dimensions : ',resized.shape)
    cv2.imshow("original", photo)
    cv2.imshow("Resized image", resized)
    if cv2.waitKey(100) ==13:
        break
cv2.destroyAllWindows()
```

```python
cap.release()
```
