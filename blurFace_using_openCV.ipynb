```python
import cv2
```

```python
cameraIndex = 'https://192.168.29.200:8080/video'
```

```python
cap = cv2.VideoCapture(cameraIndex)
model  = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
```

```python
while True:
    ret,photo = cap.read()
    fdetect = model.detectMultiScale(photo)
    if len(fdetect) == 0:
        print("no faces found")
    else:
        x1 = fdetect[0][0]
        y1 = fdetect[0][1]
        x2 = x1 + fdetect[0][2]
        y2 = y1 + fdetect[0][3]
        cv2.rectangle(photo, (x1,y1), (x2, y2), [0, 255, 0], 2)
        cphoto = photo[y1:y2, x1:x2]
        bphoto = cv2.blur(cphoto, ksize=(20, 20))
        photo[y1:y2, x1:x2] = bphoto
        cv2.imshow("hey", photo)
        if cv2.waitKey(100)==13:
            break
cv2.destroyAllWindows()
cap.release()
```
